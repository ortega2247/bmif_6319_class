# Introduction to Enzymes

```python
import numpy as np
from IPython.display import Image, Video
```

1. Enzymes are proteins that act as biological catalysts (biocatalysts).
2. An enzyme attracts substrates to its active site, catalyzes the chemical reaction by which products are formed, and then allows the products to dissociate (separate from the enzyme surface).
3. Enzymes increase the reaction rate by lowering its activation energy.
4. Enzymes are not consumed in chemical reactions.
5. Enzyme activity can be affected by other molecules: inhibitors are molecules that decrease enzyme activity, and activators are molecules that increase activity.
5. An extreme example is orotidine 5'-phosphate decarboxylase, which allows a reaction that would otherwise take millions of years to occur in milliseconds

## Running the simulation with different parameter values


```python
def compare_data_simulation(traj):
    plt.figure(figsize=(10, 6))
    plt.plot(tspan, traj.all['__s5'], label='P sim', color='b')
    plt.errorbar(tspan, product_mean, yerr=product_std, fmt='ro', label='P data')
    plt.legend(loc=0)
    plt.ylabel('Concentration')
    plt.xlabel('Time (s)')
    plt.show()
                 
```

```python
compare_data_simulation(y)
```
